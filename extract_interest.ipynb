## 1. わかち書き

```python
tweets = Tweets()
# わかち書き
tweets_wakati_series = tweets.wakati(tweets_df[TWEETS_COLUMN_NAME].copy()[1:1000], 'wakati_tweet')
# トークン化
tweets_words_series = tweets.tokenize(tweets_wakati_series.copy(), 'tweets_words')
```

```python
tweets = Tweets()
result_series = tweets_df[TWEETS_COLUMN_NAME].copy()[0:1]
cw = CompoundWords()
result_series = result_series.map(lambda x: cw.pre_filter(x))

result_series = result_series.str.split(r'\s+|"|\'|、|。|\?|!|\[|\]|\(|\)|\\|「|」')
result_series = result_series.map(lambda x: cw.wakatis(x))
```

## 2. Bag-of-Words

```python
# 辞書を作成
all_tokens = tweets_words_series.copy().tolist()
dictionary = tweets.create_dictionary(all_tokens)
# Bag-of-Words(=corpus)化
corpus = [dictionary.doc2bow(tweet) for tweet in all_tokens]
tweets_bow = corpus.copy()
tweets_bow = tweets.bow(tweets_bow)
tweets_bow_series = tweets.disp_series(tweets_bow, name='bow_score', dictionaried=True, rounded=False)
tweets_bow_series.to_csv(OUT_DIR + 'tweets_bow.txt', index=False)
```

```python
sum_tweets_bow = []
[sum_tweets_bow.extend(tweets_bow_list) for tweets_bow_list in tweets_bow_series]
sum_tweets_bow = sorted(sum_tweets_bow, key=lambda x: (-int(x[1])))
with open(OUT_DIR + 'sum_bow_score_' + target_screen_name + '.csv', 'w') as f:
    [f.write('%s,%s\n' % (w[0], w[1])) for w in sum_tweets_bow]
```

```python
from collections import Counter
counter = Counter([bow[0] for bow in sum_tweets_bow])
wakati_freq_list = [(word, cnt) for word, cnt in counter.most_common()]
wakati_freq_list = sorted(wakati_freq_list, key=lambda x: (-x[1], x[0]))
pd.DataFrame(pd.Series(wakati_freq_list))
```

```python
pd.DataFrame(pd.Series(sorted(sum_tweets_bow, key=lambda x: (-x[1], x[0]))))
```

```python
sentences = tweets_df[TWEETS_COLUMN_NAME][:1000]
named_entity_json = [Tweets.extract_named_entity(sentence) for sentence in sentences]
tweets_ne_series = pd.Series(named_entity_json)
```

```python
pd.DataFrame(tweets_ne_series)
```

```python
sum_tweets_ne = []
[sum_tweets_ne.extend(tweets_ne_list) for tweets_ne_list in tweets_ne_series]
sum_tweets_ne = sorted(sum_tweets_ne, key=lambda x: (x[1], x[0]))

from functools import reduce
import collections

TYPES = ['ART', 'DAT', 'LOC', 'ORG', 'PSN', 'TIM']

# tweets_ne = [[tweets_ne_list[0] for tweets_ne_list in sum_tweets_ne if tweets_ne_list[1] == type] for type in TYPES]
sum_tweets_ne = {type: collections.Counter([tweets_ne_list[0] for tweets_ne_list in sum_tweets_ne if tweets_ne_list[1] == type]) for type in TYPES}
sum_tweets_ne = [[(key, w, sum_tweets_ne[key][w]) for w in sum_tweets_ne[key].keys()] for key in list(sum_tweets_ne)]
sum_tweets_ne_n = []
[sum_tweets_ne_n.extend(x) for x in sum_tweets_ne]
sum_tweets_ne = sorted(sum_tweets_ne_n, key=lambda x: (x[0], -x[2], x[1]))
sum_tweets_ne

with open(OUT_DIR + 'sum_ne_score_' + target_screen_name + '.csv', 'w') as f:
    [f.write('%s, %s, %s\n' % (x[0], x[1], x[2])) for x in sum_tweets_ne]
```

> 固有表現抽出は，今まで様々な方法で行われている．その中でも大きく分けるとパターン照合による固有表現抽出と，機械学習による固有表現抽出にわけられる．


> パターン照合による固有表現抽出とは，あらかじめ人手で固有表現のパターンを作成し，合致する部分をコーパスから発見することによって行われる固有表現抽出のことである（竹本，福島，山田 (2001)）．パターンとは「さん」や「大学」などの固有表現に付属しやすい文字列を指す．しかし，ルール作成のコストが高く，そこに合致しない固有表現は抽
出できない

> 人手でパターンを作成するコストや，更新するコストを解決するために，機械学習による固有表現抽出の研究も行われている．機会学習による固有表現抽出は，学習用のコーパスを用意することで，自動で抽出パターンを学習することができる．機械による手法はSVM(Support Vector Machine)（山田，工藤，松本(2002)）を利用した抽出や文節情報を利用した抽出（中野，平井(2004)）などが存在し成果を上げている．その他にも HMM(HiddenMarkov Model)や分類機の逐次適応，CRF(Conditional Random Fields)を利用した固有表現抽出なども一般的である．機械学習の問題点としては人手によるコーパス作成（橋本，乾，村上(2008)）のコストが高いことなどがある．
